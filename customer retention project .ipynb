```python
plt.figure(figsize=(15,12))
sn.countplot(df["3 Which city do you shop online from?"])
plt.title("3 Which city do you shop online from?")
plt.show()
```

The above graph shows that highest number of customers are from Delhi and the lowest number of customers are from Bulandshahr

```python
plt.figure(figsize=(15,12))
sn.countplot(df["4 What is the Pin Code of where you shop online from?"])
plt.title("4 What is the Pin Code of where you shop online from?")
plt.show()
```

```python
plt.figure(figsize=(15,12))
sn.countplot(df["5 Since How Long You are Shopping Online ?"])
plt.title("5 Since How Long You are Shopping Online ?")
plt.show()
```

The above graph shows that highest number of customer are shopping for more than 4 years and least number of customer are shopping from 1-2 years 

```python
plt.figure(figsize=(15,12))
sn.countplot(df["6 How many times you have made an online purchase in the past 1 year?"])
plt.title("6 How many times you have made an online purchase in the past 1 year?")
plt.show()
```

The above graph shows that highest number of customers have shop for less than 10 times and least number of customers have shop for 42 times and above

```python
plt.figure(figsize=(15,12))
sn.countplot(df["7 How do you access the internet while shopping on-line?"])
plt.title("7 How do you access the internet while shopping on-line?")
plt.show()
```

The above graph shows that highest number of customers use mobile internet for shopping online and least number of customers use Dial-up for shopping online 

```python
plt.figure(figsize=(15,12))
sn.countplot(df["8 Which device do you use to access the online shopping?"])
plt.title("8 Which device do you use to access the online shopping?")
plt.show()
```

The above graph shows that highest number of customers use smartphone for shopping online and least number of customers use laptop for shopping online

```python
plt.figure(figsize=(15,12))
sn.countplot(df["18 The content on the website must be easy to read and understand"])
plt.title("The content on the website must be easy to read and understand")
plt.show()
```

The above graph shows that highest number of customers strongly agree that the content on the website must be easy to read and understand  and least number of customers feel indifferent.
