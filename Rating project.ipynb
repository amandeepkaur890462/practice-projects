```python
#importing all the required libraries
import selenium
import pandas as pd
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
```

```python
#connecting to the web drivers
driver= webdriver.Chrome(r"C:\chromedriver.exe")
```

```python
driver.get('https://www.amazon.in/')
```

```python
Ratings=[]
Titles=[]
```

```python
search_bar= driver.find_element_by_id('twotabsearchtextbox')
search_bar.clear()
search_bar.send_keys("laptops")
search_button=driver.find_element_by_xpath('//span[@id="nav-search-submit-text"]').click()
```

```python
urls=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
UR=[]
for i in urls[:100]:
    UR.append(i.get_attribute('href'))
for url in UR:
    driver.get(url)
    try:
        rate=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        rate.click()
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Ratings.append(rating.text)
        title=driver.find_element_by_xpath("//span[@class='a-size-large product-title-word-break']")
        title.click()
        Titles.append(title.text)
        
    except NoSuchElementException as e:
        Ratings.append("No rating")  
```
